### Data Loading

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

df = pd.read_csv('loan_data_2007_2014.csv', index_col=[0], low_memory=False)

# dropping irelevant column
df = df.drop(['id'], axis=1)
df.head()
```

```python
df.info()
```

### Dropping Column with 0 Values

```python
df = df.drop([
    'annual_inc_joint', 
    'dti_joint', 
    'verification_status_joint', 
    'open_acc_6m',
    'open_il_6m',
    'open_il_12m',
    'open_il_24m',
    'mths_since_rcnt_il',
    'total_bal_il',
    'il_util',
    'open_rv_12m',
    'open_rv_24m',
    'max_bal_bc',
    'all_util',
    'inq_fi',
    'total_cu_tl',
    'inq_last_12m'
], axis=1)

df.head()
```

## Visualize Business Performance

```python
performance = df.loc[:, ['loan_amnt', 'funded_amnt', 'term', 'issue_d']]
performance.head()
```

### Assume 36 months are Short Term loan

```python
short_term = performance.loc[performance['term'] == ' 36 months', :].drop(['term'], axis=1)
short_term['issue_m'] = short_term['issue_d'].str[:3]
short_term.head()
```

```python
ordered_month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
short_term_groupby_month = short_term.groupby('issue_m')[['loan_amnt', 'funded_amnt']].sum()
short_term_groupby_month = short_term_groupby_month.reindex(ordered_month)
short_term_groupby_month
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

fig, ax = plt.subplots(1, 2, figsize=(20,7))
sns.lineplot(
    data=short_term_groupby_month, 
    x=short_term_groupby_month.index, 
    y='loan_amnt', 
    ax=ax[0]
)
ax[0].ticklabel_format(style='plain', axis='y')
ax[0].tick_params(labelrotation=90, axis='x')
ax[0].set_title('Short Term Loan Given for Each Month', fontsize=14, pad=20)

sns.lineplot(
    data=short_term_groupby_month, 
    x=short_term_groupby_month.index, 
    y='funded_amnt', 
    ax=ax[1]
)
ax[1].ticklabel_format(style='plain', axis='y')
ax[1].tick_params(labelrotation=90, axis='x')
ax[1].set_title('Short Term Fund Received Each Month', fontsize=14, pad=20)
fig.tight_layout()
plt.show()
```

### Calculate Percentage for Loan Amount at the End of the Year

```python
loan_amount_end = short_term_groupby_month.loc[short_term_groupby_month.index.values[-1], 'loan_amnt']
loan_amount_start = short_term_groupby_month.loc[short_term_groupby_month.index.values[0], 'loan_amnt']
short_loan_amount_percentage = (loan_amount_end - loan_amount_start) / loan_amount_start * 100
print('The short term loan amount at the end of year is increasing by {:.2f}%'.format(short_loan_amount_percentage))
```

### Calculate Percentage for Funded Amount at the End of the Year

```python
fund_amount_end = short_term_groupby_month.loc[short_term_groupby_month.index.values[-1], 'funded_amnt']
fund_amount_start = short_term_groupby_month.loc[short_term_groupby_month.index.values[0], 'funded_amnt']
short_fund_amount_percentage = (fund_amount_end - fund_amount_start) / fund_amount_start * 100
print('The short term funded amount at the end of year is increasing by {:.2f}%'.format(short_fund_amount_percentage))
```

```python
unpaid_short_loan = short_loan_amount_percentage - short_fund_amount_percentage
print('The unpaid short term loan at the end of the year is {:.2f}%'.format(unpaid_short_loan))
```

### Assume 60 months are Long Term Loan

```python
long_term_groupby_month = long_term.groupby('issue_m')[['loan_amnt', 'funded_amnt']].sum()
long_term_groupby_month = long_term_groupby_month.reindex(ordered_month)
long_term_groupby_month
```

### Calculate Percentage for Loan Amount at the End of the Year

```python
loan_amount_end = long_term_groupby_month.loc[long_term_groupby_month.index.values[-1], 'loan_amnt']
loan_amount_start = long_term_groupby_month.loc[long_term_groupby_month.index.values[0], 'loan_amnt']
long_loan_amount_percentage = (loan_amount_end - loan_amount_start) / loan_amount_start * 100
print('The long term loan amount at the end of year is increasing by {:.2f}%'.format(long_loan_amount_percentage))
```

### Calculate Percentage for Funded Amount at the End of the Year

```python
fund_amount_end = long_term_groupby_month.loc[long_term_groupby_month.index.values[-1], 'funded_amnt']
fund_amount_start = long_term_groupby_month.loc[long_term_groupby_month.index.values[0], 'funded_amnt']
long_fund_amount_percentage = (fund_amount_end - fund_amount_start) / fund_amount_start * 100
print('The short term funded amount at the end of year is increasing by {:.2f}%'.format(long_fund_amount_percentage))
```

```python
unpaid_long_loan = long_loan_amount_percentage - long_fund_amount_percentage
print('The unpaid long term loan at the end of the year is {:.2f}%'.format(unpaid_long_loan))
```

```python
average_loan = (short_loan_amount_percentage + long_loan_amount_percentage) / 2
average_fund = (short_fund_amount_percentage + long_fund_amount_percentage) / 2
print('Average loan at the end of the year is {:.2f}%'.format(average_loan))
print('Average paid loan at the end of the year is {:.2f}%'.format(average_fund))
```
